```python
import numpy as np

num_session = 100
packages_per_session = 50
tokens_per_package = 8

dataset = np.random.rand(num_session, packages_per_session, tokens_per_package)
print(dataset.shape)
```

```python
marks = np.random.uniform(0.1, 1.1, size=(num_session, packages_per_session, 1))
print(marks.shape)
```

```python
import keras

model = keras.Sequential([
    keras.layers.Input(shape=(None, 8)),
    keras.layers.LSTM(64, return_sequences=True),
    keras.layers.LSTM(64, activation="relu", return_sequences=True),
    keras.layers.Dense(1, activation="sigmoid")
])

optimizer = keras.optimizers.Adam(learning_rate=0.001)
loss = keras.losses.BinaryCrossentropy()
metrics = [keras.metrics.Accuracy, keras.metrics.Precision, keras.metrics.Recall, keras.metrics.AUC]
model.compile(optimizer=optimizer, loss=loss, metrics=metrics)
```

```python
model.summary()
```

```python
model.fit(x=dataset, y=marks, epochs=10, batch_size=32)
```
